## How to detect RNA modifications by nanopore sequencing 

**Direct basecalling of modified nucleotides from the raw signal**
* Based on Deep Learning Basecalling
* Training from a labeled raw signal with all the possible kmers combinations 
* Can potentially identify a large number of modified bases simultaneously

**Post-processing of the basecalled signal**
* Identification of modified bases by comparison between non-modified and modified samples
* Much simpler approach
* Rely on the basecaller efficiency

## Training with Chiron

**Necessary code modifications**
* Debugging
* 5 bases instead of 4

**Test with DNA data**
&#10137; Works fine

**RNA: 5 times slower than DNA &#10137; Need to tune the hyperparameters**
* Batch size
* Step rate
* Max number of steps
* Sequence length
* Sliding step size (basecalling)
